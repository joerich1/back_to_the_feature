```python
import os

cwd = os.getcwd()
entries = os.listdir(cwd)
json_files = [
            entry for entry in entries
            if os.path.isfile(os.path.join(cwd, entry)) and entry.lower().endswith('.json')
        ]
json_files
```

```python
import json

keepers = ["budget", "genres", "origin_country", "overview", "release_date", "revenue", "runtime", "tagline",
          "title", "vote_average", "vote_count", "credits", "actors_characters"]
movies_dict = {}

for file_ind in range(len(json_files)):
    try:    
        data = json.load(open(json_files[file_ind], encoding='utf-8'))
        if data["revenue"] >10_000_000:
            movies_dict[data["title"]] = {k: data[k] for k in keepers}
    except:
        print(json_files[file_ind])
```
