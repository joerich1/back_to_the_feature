```python
data_dir = "data/movie_data"
filename = "data/movie_data/Back_To_The_Future.json"

import json

# Load JSON file
with open(filename, "r", encoding="utf-8") as file:
    data = json.load(file)

# Print first few entries to check
print(list(data.items())[:10])  # Prints first 10 key-value pairs
```

```python
import re

characters = [dicty["character"] for dicty in data["actors_characters"]]
characters_cleaned = []
for char in characters:
     names = char.split(" ")
     for name in names: 
         name = re.sub(r"[^a-z]", "", name.lower())
         if name != "":
             characters_cleaned.append(name)

characters_cleaned
```
